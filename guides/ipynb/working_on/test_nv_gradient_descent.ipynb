```python
from tensorflow import keras
from keras.datasets import mnist
import numpy as np
```

```python
input_dim = 784
original_inputs = keras.Input(shape = input_dim,)
base_model = keras.layers.Dense(64, activation=keras.activations.relu)(original_inputs)
base_model = keras.layers.Dense(10, activation=keras.activations.softmax)(base_model)
```

```python
(x_train, y_train), (x_test, y_test) =  mnist.load_data()
x_train.shape
```

```python
x_train = x_train.reshape((x_train.shape[0], np.prod(x_train.shape[1:]))).astype("float32")/255
x_train[0:3,400:410]
```

```python
model = keras.Model(inputs=original_inputs, outputs=base_model)
model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
model.summary()
```

```python
model.fit(x_train, y_train, batch_size=32, epochs=3)
```
